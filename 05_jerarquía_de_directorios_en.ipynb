# *GNU/Linux* directory hierarchy.

*UNIX* and *GNU/Linux* systems have a standardized directory structure, which has been proposed by the *Linux Foundation*.

This specification is known as the standard file system hierarchy ([*fhs*](https://wiki.linuxfoundation.org/lsb/fhs)).

The current specification can be consulted from:



https://refspecs.linuxfoundation.org/FHS_3.0/fhs/index.html

### The root directory ```/```.

This is the home directory of a *UNIX* or *GNU/Linux* system. It is used as the reference to the rest of the system.

````bash
tree / -L 1
````

### The ```/home/``` directory.

This is the directory where the *home* of each user on the system is located.

````bash
tree /home -L 1
````

### The ```/bin/``` directory.

It is the directory where the main system commands are located.

````bash
tree /bin -L 1
````

### ### The ```/sbin/``` directory.

It is the directory of "safe" applications.

````bash
tree /sbin -L 1
````

### The ```/lib/``` and ```/lib64/``` directories.

This directory contains the system libraries.

````bash
tree /lib/ -L 1
````

````bash
tree /lib64/ -L 1
````

### The ```/lib/modules/``` directory.

This directory contains the *Linux* kernel module libraries.

````bash
tree /lib/modules -L 2
````

### The ```/etc/``` directory.

This is the directory that contains the directories and configuration files of the system and its components.

````bash
tree /etc -L 2
````

````bash
tree /etc/apache2
````

### The ```/tmp/``` directory.

It is the directory that holds temporary files for the entire system and all users.

````bash
tree -L 1 /tmp
````

### The ```/var/``` directory.

This directory contains files that are used by applications and are not owned by a regular user.

````bash
tree /var/ -L 2
````

### The ```/var/log/``` directory.

This directory contains the system logs.

````bash
tree /var/log -L 1
````

### The ```/var/www/``` directory

This is the default directory for the Apache root.

````bash
tree /var/www -L 2
````

### The ```/usr/``` directory

This directory contains access (execution) directories and files for all users.

````bash
tree /usr/ -L 1
````

### The ```/usr/local/``` directory.

It is a directory that contains resources that are not for general use by the system, but can be used by specific groups of users.

````bash
tree /usr/local/ -L 1
````

### The ```/opt/``` directory.

This directory is for various applications.

````bash
tree /opt/
````

### The ```/boot/``` directory.

This is the directory where the * Linux * kernel image and system boot configuration is located.

````bash
tree //boot/
````

### Directory ```/dev/```.

The directory that links to the devices detected and configured by the kernel drivers.

````bash
tree /dev -L 1
````

### The ```/proc/``` directory.

This is the directory that maps the system processes.

````bash
tree /proc/ -L 1
````

### The ```/sys/``` directory.

This is the directory used by the kernel libraries.

````bash
tree /sys/ -L 1
````

### The ```/run/``` directory.

It is the directory that contains execution files of certain services and daemons.

````bash
tree /run/ -L 1
````

### The ```/media/``` and ```/mnt/``` directories.

These are the directories in which the various file systems are mounted.

````bash
tree /mnt/ -L 2
````

````bash
tree /media/ -L 2
````

## The *PATH* environment variable.

This variable contains the current user's command search paths.

````bash
echo $PATH
````
