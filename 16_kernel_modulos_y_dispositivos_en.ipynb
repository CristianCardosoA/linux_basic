## Linux kernel files.

The Linux kernel architecture allows for a "monolithic" component to which modules can be added dynamically.

### The monolithic kernel component.

The monolithic, kernel component contains the critical elements of a system and is typically an executable file whose name starts with [*vmlinuz*](http://www.linfo.org/vmlinuz.html) and is located in the system's */boot/* directory.

````bash
ls /boot/vmlinuz*
````

## The ```uname``` command.

That command returns basic system information, including the current kernel version.

```
uname <opciones>
```
### Some ```uname``` options.

* ```-a``` or ```--all``` returns all system data.
* ```-s``` or ```--kernel-name``` returns the name of the kernel.
* ```-r``` or ```--kernel-release``` returns the Linux kernel version.
* ```-v``` or ```--kernel-version``` returns the version of the kernel build.
* ```-m``` or ```--machine``` returns data for the processor it was compiled for.
* ```-o```o ```--operating-system``` returns all data for the operating system it was compiled for.

````bash
uname
````

````bash
uname -a
````

````bash
uname -s
````

````bash
uname -r
````

````bash
uname -v
````

````bash
uname -m
````

````bash
uname -o
````

### The ```uname``` man page.

````bash
man uname
````

### The kernel modules.

In addition to the monolithic component, the system can add "modules" to the kernel, which are in the ```/lib/modules``` directory. Typically, the directory it is in corresponds to ```uname -r```.

````bash
ls /lib/modules/$(uname -r)
````
