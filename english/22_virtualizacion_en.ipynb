# Basic concepts.

### Virtualization.

* Consists of the management and use of both software and hardware resources with the objective of creating representations of equipment, servers, networks and computing infrastructure capable of providing identical functionalities to those of their physical counterparts.

### Universal machines.
* The mathematician Alan Turing defined a universal machine as a system capable of completely emulating the functions of other universal machines.
* Virtualization is based on the principle of universal Turing machines.

### Types of virtualization.
* Machines (VM, Containers, Zones, etc.).
* Redes (SDN, NFV).
* Storage (SDS).
* Applications.

* The system that contains and manages the virtualized elements is known as _Host_ (*Host*).
* Virtualized systems within the Host are known as _Guests_ (*Guest*).

# Virtual Machines (VMs).

Virtual machines are systems capable of simulating in a simplified but effective way the components (memory, processors, input and output devices, etc.) of a physical machine.

#### Complete virtualization.
It is when it is possible to emulate from the host all the functionalities in such a way that the host system can be executed exactly in the same way as in a physical system.
#### Hardware virtualization.
Processor manufacturers such as Intel (VT) and AMD (V) have included technologies within their products that make it easy for host systems to deliver optimized virtualized environments from the hardware itself.

#### Paravirtualization.
It is when it is required to make certain modifications in the host system to be able to work in the host.

# Hypervisors or virtual machine monitors.

* The component in charge of managing the resources and the execution of the virtual machines is known as the hypervisor.
* Hypervisors can be contained within the host operating system or even be a component external to the host operating system.

## Network virtualization.

###  [Linux Bridge.](https://wiki.linuxfoundation.org/networking/bridge)
Linux Bridge is a component of the Linux kernel that allows virtualizing the functions of a bridge in a simplified way.

### Software-Defined Networks (SDN).

#### Plans.
* Data. In which user data is transmitted.
* Control. In which network configuration data is transmitted.
* Administration. The one in charge of network management.

#### The OpenFlow protocol.
The open protocol used for SDN implementation is [OpenFlow](https://www.opennetworking.org/).
Some implementations of the protocol are:
* [Open VSwitch.](http://openvswitch.org/)
* [Open Daylight.](https://www.opendaylight.org/)
* [open Contrail.](http://www.opencontrail.org/)
* [Mini Net.](http://mininet.org/)
* [Floodlght](http://www.projectfloodlight.org/floodlight/)
