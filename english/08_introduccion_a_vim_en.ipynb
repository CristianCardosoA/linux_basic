## The ```vi``` and ```vim``` editors.

There are several text editor options for *UNIX* and *GNU/Linux*, however, all include the [*vi*](https://en.wikipedia.org/wiki/Vi) editor or more modern adaptations such as [*vi*](https://en.wikipedia.org/wiki/Vi) or [*vi*](https://en.wikipedia.org/wiki/Vi).

The ```vi``` family of editors are very popular and revered due to their wide range of commands, options and extensions, but also one of the most feared, due to the fact that they require the user to memorize various key combinations and faced with a text interface that at first is not very friendly for users accustomed to text editors in graphical environments.

The *vim* editor has multiple extensions for programming languages ​​and software development.

It is possible to consult a list of *vim* commands from https://vim.rtorr.com/lang/es_es

## *vim* modes.

One of the main features of ```vi``` is that it has two modes:

* The command mode.
* Insert mode.

### Command mode.

Command mode allows you to execute various instructions that the text editor executes.

To access command mode from edit mode, you must type the <kbd>ESC</kbd> key.

### Edit mode.

Edit mode allows you to scroll through the text and modify it.

The following commands allow you to enter edit mode.
* <kbd>i</kbd> edit mode.
* <kbd>or</kbd> edit mode leaving a line return.
* <kbd>I</kbd> edit mode at the beginning of the line.
* <kbd>The</kbd> top line editing mode.

## Offset.

* <kbd>h</kbd> moves the cursor to the left.
* <kbd>j</kbd> moves the cursor down.
* <kbd>k</kbd> moves the cursor up.
* <kbd>l</kbd> moves the cursor to the right.

### Search and replace.

The ```vi```editor allows you to perform advanced search operations that can include regular expressions.

#### Simple search.

<kbd>:</kbd><kbd>/</kbd>```<search pattern>```

Where:

* ```<search pattern>``` is a string or regular expression.

#### Find and replace a pattern.

<kbd>:</kbd><kbd>s</kbd><kbd>/</kbd>```<search chain>```<kbd>/</kbd>```<replacement chain >```

Where:

* ```<search pattern>``` is a string or regular expression.
* ```<pattern substitution>``` is the character string
which will replace the first string matching ```<search pattern>```.

#### Find and replace all matches.

<kbd>:</kbd><kbd>%</kbd><kbd>s</kbd><kbd>/</kbd>```<search pattern>```<kbd>/</kbd >```<replacement chain>```

Where:

* ```<search pattern>``` is a string or regular expression.
* ```<replacement string>``` is the character string
which will substitute for strings matching ```<search pattern>```.

### Copy, cut and paste.

#### Select.

* <kbd>v</kbd> Starts character selection mode. Once started, the user will mark the block that will be selected character by character.
* <kbd>V</kbd> Starts column selection mode. Once it starts, the user will mark the lines that will be selected.
* <kbd>Ctrl</kbd><kbd>v</kbd> Starts section selection mode. Once started, the user will be able to define a column of a specific width along several lines that will be selected.


#### Copy.

* <kbd>y</kbd> This command copies the selected block of text.

#### Cut.

* <kbd>d</kbd> This command pastes the selected block of text.
