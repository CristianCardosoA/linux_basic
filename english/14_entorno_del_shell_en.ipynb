# Enter the shell with *Bash*.

## The ```/etc/skel``` directory.

This directory contains the environment configuration files that will be used by default when creating a new user.

````bash
ls -a /etc/skel
````

### Environment configuration files.

These files are located in the user's home and are automatically executed each time a user accesses the system.

* ```~/.bashrc```
* ```~/.profile```
* ```~/.bash_logout```
* ```~/.bash_history```


````bash
cat ~/.bashrc
````

````bash
cat ~/.profile
````

````bash
cat ~/.bash_logout
````

````bash
cat ~/.bash_history
````

## Environment Variables.

An environment variable is defined for all of a user's processes in a session.

Environment variables are defined by the ```export``` command.

### The ```export``` command.

If no arguments are added, ```export``` will return the list of environment variables.

**Examples:**

````bash
export
````

````bash
export NOMBRE=Jose
````

````bash
export LANG="es_MX"
````

````bash
export DEITY=Ala
````

````bash
export | egrep -i '(ala|jose|es_MX)'
````
